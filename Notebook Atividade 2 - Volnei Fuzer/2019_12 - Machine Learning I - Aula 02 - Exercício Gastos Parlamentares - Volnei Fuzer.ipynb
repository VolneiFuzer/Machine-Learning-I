```python
# Avaliando a existencia de valores nulos

print(gastos_parlamentares.isnull().sum())
```

```python
# Apesar de não existir valores nulos, foi possível observar pelos primeiros registros
# que existem muitos campos com o caracter "-" assim para avaliar os casos realemtne sem informação vamos substituir

gp_tratamento = gastos_parlamentares.apply(lambda x: x.replace('-',np.NaN))
gp_tratamento.head(100)
```

```python
# realizando a avaliação dos valores nulos

gp_tratado = gp_tratamento.isna().sum().sort_values(ascending=False)
perc_1 = gp_tratamento.isna().sum()/gp_tratamento.isna().count()*100
perc_2 = perc_1.sort_values(ascending=False)
dados_fatantes = pd.concat([gp_tratado, perc_2], axis=1, keys=['Total', '%'])
dados_fatantes.head(12)
```

```python
# Removendo as colunas que não possuem dados suficientes para avaliação

gp_tratamento = gp_tratamento.drop(columns=['txtPassageiro', 'txtTrecho', 'txtDescricaoEspecificacao', 'txtCNPJCPF', 
                                                'txtNumero', 'datEmissao'], axis=1)
gp_tratamento.head()
```

```python
# Removendo as linhas de registros em que poucas linhas são nulas

gp_tratamento = gp_tratamento.dropna(subset=['codLegislatura', 'ideCadastro', 'nuCarteiraParlamentar',
                                               'sgUF', 'sgPartido'])
```

```python
gp_tratado = gp_tratamento

gp_filtro = gp_tratado.iloc()

#gp_filtro = gp_tratado[['vlrLiquido', 'vlrLiquido']]

#gp_filtro.head()
```

```python
dendrograma = sch.dendrogram(sch.linkage(gp_filtro, method='ward'))
plt.title('Dendrograma')
plt.xlabel('Valor Líquido')
plt.ylabel('Valor Líquido')
plt.show()
```
